```python
# import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy 
import requests
import json
from api_keys import weather_api_key
from citipy import citipy 
import random
from datetime import date
import scipy.stats as st
weather_date = str(date.today())

```

```python
#get random citiets
cities = []
while len(cities) < 1000:
    city = citipy.nearest_city(random.uniform(-90,90),random.uniform(-180,180))
    cities.append(city.city_name)
    #eliminate duplicates
    cities = list(set(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
#empty lists to add values per city to, and then make df
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
iteration = 0
#loop thru all the cities and print each as data is retreived
for city in cities: 
    try:
        query_url = f"{url}appid={weather_api_key}&q={city}"
        responses = (requests.get(query_url).json())
        iteration += 1
        print(f"getting weather data for {city}(city #{iteration}) ")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        max_temp.append((responses['main']['temp_max'])-273.15)
        lat.append(responses['coord']['lat'])
        lng.append(responses['coord']['lon'])
        humidity.append(responses['main']['humidity'])
        cloudiness.append(responses['clouds']['all'])
        wind_speed.append(responses['wind']['speed'])
        country.append(responses['sys']['country'])
        date.append(responses['dt'])
        
#if city isnt in api skip it and go to next, append lists with null so list values correspond to cities        
    except:
        KeyError
        print(f'KeyError - {city} Skipped --- next city')
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        max_temp.append(None)
        lat.append(None)
        lng.append(None)
        humidity.append(None)
        cloudiness.append(None)
        wind_speed.append(None)
        country.append(None)
        date.append(None)
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#put lists with data values into a dict...
weather_dict = {
    "City": cities,
    "lat": lat,
    "lng": lng,
    "max temp":max_temp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_speed,
    "Country": country,
    "Date": date
}
#...and make a dataframe out of it
weather_df = pd.DataFrame(weather_dict)
weather_df = weather_df.dropna()
#export to csv
weather_df.to_csv("city_weather_csv")
weather_df

```

```python
#get the quantitative values from weather df to make new df to see that there arent any errors (eg. humidity > 100)
weather_df_numeric = weather_df[['lat','lng','max temp','Humidity','Cloudiness','Wind Speed','Date']]
#make list of functions to loop thru columns
function_list = ['count', 'mean', 'std', 'min', 'max']
#and a list of quantiles
quantiles = [.25,.5,.75]


weather_stats_df = weather_df_numeric.agg({"lat" : function_list})

for col in weather_df_numeric:
    col_stats_df = weather_df_numeric.agg({col : function_list})
    weather_stats_df[col] = col_stats_df

for q in quantiles:
    weather_stats_df = weather_stats_df.append(weather_stats_df.quantile(q= q))
weather_stats_df
```

## Latitude vs. Temperature Plot

```python
def plot_weather(y_axis):
    x = weather_df[['lat']]
    y = weather_df[[y_axis]]
    lat_vs_humidity_plot = plt.scatter(x,y,edgecolors = "black")
    plt.title(f"Latitude and {y_axis} per city on {weather_date}")
    plt.xlabel("Latitude")
    plt.ylabel(y_axis)
    plt.grid(True)
    lat_vs_humidity_plot
    plt.savefig(f"Lat_{y_axis}.png")
plot_weather("max temp")
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
def hemisphere_regression(hemisphere, y_axis):

    if hemisphere == 'north':
        df = weather_df[weather_df["lat"] > 0]
    elif hemisphere == "south":
        df = weather_df[weather_df["lat"] < 0]
    else:
        print('argument 0 must be /"north" or /"south"')
    x = df['lat']
    y = df[y_axis]
    plt.scatter(x,y,edgecolors = "black")
    plt.xlabel("Latitude")
    plt.ylabel(y_axis)
    plt.title(f"Latitude and {y_axis} per city on {weather_date} in the {hemisphere}ern hemishpere")
    slope, intercept, r, p, sem = st.linregress(x, y)
    plt.grid(True)

    fit = slope * x + intercept
    plt.plot(x, fit, color ="r")

    corr = round(st.pearsonr(x,y)[0],9)

    print(f'coorelation = {corr}')

    plt.savefig(f"{hemisphere}_{y_axis}.png")
    
hemisphere_regression('north','max temp')
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
hemisphere_regression("south","max temp")
```

```python
#Based on the high correlation coefficient of temperature and latitude for both the north and south hemispheres, 
# it seems that the temperature tends to increase as latitude approches zero 
```
