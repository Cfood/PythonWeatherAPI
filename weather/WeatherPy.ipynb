```python
# import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy 
import requests
import json
from api_keys import weather_api_key
from citipy import citipy 
import random
from datetime import date
import scipy.stats as st
weather_date = str(date.today())

```

```python
#get random citiets
cities = []
while len(cities) < 1000:
    city = citipy.nearest_city(random.uniform(-90,90),random.uniform(-180,180))
    cities.append(city.city_name)
    #eliminate duplicates
    cities = list(set(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
iteration = 0
for city in cities: 
    try:
        query_url = f"{url}appid={weather_api_key}&q={city}"
        responses = (requests.get(query_url).json())
        iteration += 1
        print(f"getting weather data for {city}(city #{iteration}) ")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        max_temp.append((responses['main']['temp_max'])-273.15)
        lat.append(responses['coord']['lat'])
        lng.append(responses['coord']['lon'])
        humidity.append(responses['main']['humidity'])
        cloudiness.append(responses['clouds']['all'])
        wind_speed.append(responses['wind']['speed'])
        country.append(responses['sys']['country'])
        date.append(responses['dt'])
        
        
    except:
        KeyError
        print(f'KeyError - {city} Skipped --- next city')
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        max_temp.append(None)
        lat.append(None)
        lng.append(None)
        humidity.append(None)
        cloudiness.append(None)
        wind_speed.append(None)
        country.append(None)
        date.append(None)
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dict = {
    "City": cities,
    "lat": lat,
    "lng": lng,
    "max temp":max_temp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_speed,
    "Country": country,
    "Date": date
}

weather_df = pd.DataFrame(weather_dict)
weather_df = weather_df.dropna()
weather_df.to_csv("city_weather_csv")
weather_df

```

```python
weather_df_numeric = weather_df[['lat','lng','max temp','Humidity','Cloudiness','Wind Speed','Date']]
function_list = ['count', 'mean', 'std', 'min', 'max']
quantiles = [.25,.5,.75]
weather_stats_df = weather_df_numeric.agg({"lat" : function_list})

for col in weather_df_numeric:
    col_stats_df = weather_df_numeric.agg({col : function_list})
    weather_stats_df[col] = col_stats_df

for q in quantiles:
    weather_stats_df = weather_stats_df.append(weather_stats_df.quantile(q= q))
weather_stats_df
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
north_df = weather_df[weather_df["lat"] > 0]

x_north = north_df['lat']
y_north = north_df['max temp']
plt.scatter(x_north,y_north,edgecolors = "black")
plt.xlabel("Latitude")
plt.ylabel("max temp")
plt.title(f"Latitude and Max Temp per city on {weather_date}")
lat_vs_temp_plot
slope, intercept, r, p, sem = st.linregress(x_north, y_north)
plt.grid(True)

fit = slope * x_north + intercept
plt.plot(x_north, fit, color ="r")

corr = round(st.pearsonr(x_north,y_north)[0],9)

print(f'coorelation = {corr}')

plt.savefig("north_lat_temp.png")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
south_df = weather_df[weather_df["lat"] < 0]

x_south = south_df['lat']
y_south = south_df['max temp']
plt.scatter(x_south,y_south,edgecolors = "black")
plt.xlabel("Latitude")
plt.ylabel("max temp")
plt.title(f"Latitude and Max Temp per city on {weather_date}")
lat_vs_temp_plot
slope, intercept, r, p, sem = st.linregress(x_south, y_south)
plt.grid(True)

fit = slope * x_south + intercept
plt.plot(x_south, fit, color ="r")

corr = round(st.pearsonr(x_south,y_south)[0],9)

print(f'coorelation = {corr}')

plt.savefig("south_lat_temp.png")
```
